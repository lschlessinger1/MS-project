# Parent selection probability distribution

```python
%load_ext autoreload
%autoreload 2
```

```python
from src.evalg.selection import ExponentialRankingSelector, LinearRankingSelector

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(25)
raw_fitness_values = -np.arange(25) + 100
```

```python
lr_selector = LinearRankingSelector(25)
er_selector = ExponentialRankingSelector(25, c=0.7)
```

```python
lr_prob = lr_selector.get_probabilities(raw_fitness_values)
er_prob = er_selector.get_probabilities(raw_fitness_values)
```

```python
normlized_fitness = raw_fitness_values / raw_fitness_values.sum()
```

```python
plt.title('Parent Selection probability distribution')
plt.scatter(x, normlized_fitness, label='Normalized Fitness')
plt.bar(x, lr_prob, label='Linear', alpha=0.7)
plt.bar(x, er_prob, alpha=0.4, label='Exponential')
plt.xlabel('Parent fitness ranking')
plt.ylabel('Probability')
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Only show ticks on the left and bottom spines
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')

plt.legend()
# Uncomment to save figure
# plt.savefig("parent_fitness_distribution.svg")
plt.show()
```
